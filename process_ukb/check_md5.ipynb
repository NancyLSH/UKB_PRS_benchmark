```python
import pandas as pd
import numpy as np 
import sys
```

```python
md5_target = pd.read_csv("eur_md5_target.txt", sep=" ", header=None, names=["md5", "file"])
md5_source = pd.read_csv("eur_md5_source.txt", sep=" ", header=None, names=["md5", "file"])

merged = pd.merge(md5_target, md5_source, on="file", how="inner", suffixes=("_target", "_source"))
display(merged)
# check if md5_target and md5_source are the same
if np.array_equal(merged["md5_target"], merged["md5_source"]):
    print("All md5 hashes match.")
```
