```python
import pandas as pd
import numpy as np
import os
import sys
```

```python
fam_data = pd.read_csv("../../data/genotype/eur_chr22.fam", delim_whitespace=True, header=None)
fam_data.columns = ["FID", "IID", "PID", "MID", "Sex", "Pheno"]
display(fam_data.head())
```

```python
pheno_data = pd.read_csv("../../data/pheno/fields_group_1.tsv", delim_whitespace=True, header=0, low_memory=False)
display(pheno_data.head())
```

```python
pheno_data_2 = pd.read_csv("../../data/pheno/fields_group_2.tsv", delim_whitespace=True, header=0, low_memory=False)
display(pheno_data.head())
```

```python
merge_data = pd.merge(fam_data, pheno_data, left_on="IID", right_on="eid", how="left")
merge_data = pd.merge(merge_data, pheno_data_2, left_on="IID", right_on="eid", how="left", suffixes=("", "_2"))
# check if there are any IID in fam_data that are not in pheno_data
missing_iids = merge_data[merge_data["eid"].isnull()]["IID"].unique()
display(merge_data.head())
print(merge_data.columns)
primary_columns = ["p31", "p34", "p47_i0", "p48_i0", "p46_i0"]
primary_data = merge_data[["FID", "IID"] + primary_columns].copy()
# check if there are any missing values in primary_data
missing_values = primary_data.isnull().sum()
print(missing_values[missing_values > 0])
if len(missing_iids) > 0:
    print(f"Missing IID in pheno_data: {missing_iids}")
```
